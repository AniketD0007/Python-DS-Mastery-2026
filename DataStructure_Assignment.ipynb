```python
'''Q1. Write a code to reverse a string'''

string = input("Enter any sentence : ")
print("The reverse of your given statement is : ",string[::-1])
```

```python
'''Q2.  Write a code to count the number of vowels in a string'''

vowels=['a','e','i','o','u']
string = input("Enter any word that contains vowels (preferrably a long one) : ")
count=0
for i in string:
    if i in vowels:
        count+=1
print("The number of vowels in your word is : ",count)        
```

```python
'''  BETTER VERSION 1  '''

vowels='aeiou'
string = input("Enter any word that contains vowels (preferrably a long one) : ").lower()
count=len([char for char in string if char in vowels])

print(f"The number of vowels in your word is : {count}")        
```

```python
'''  BETTER VERSION 2  '''

vowels='aeiou'
string = input("Enter any word that contains vowels (preferrably a long one) : ").lower()
count=sum(char in vowels for char in string)

print(f"The number of vowels in your word is : {count}")        
```

```python
'''  BEST VERSION 3 (Removes duplicacies)  '''

vowels=set('aeiou')
string = set(input("Enter any word that contains vowels (preferrably a long one) : ").lower())


print(f"The number of vowels in your word is : {len(vowels & string)}")        
```

```python
'''  Q3. Write a code to check if a given string is a palindrome or not  '''

string=input("Enter any word : ")
if string.lower()==string[::-1].lower():
    print("The word is a palindrome")
else :   
    print("Not palindrome")
```

```python
'''  Q4.  Write a code to check if two given strings are anagrams of each other  '''

string1=input("Enter any word : ").lower()
string2=input("Enter the other word : ").lower()

print("The words are anagrams of each other" if (sorted(string1)==sorted(string2)) else "Not anagrams")
```

```python
''' Q5. Write a code to find all occurrences of a given substring within another string (asks for the positions where the substring occurs) '''


target=input("Enter the substring : ").lower()
main=input("Enter the main string : ").lower()

positions=[i for i in range(len(main)) if main.startswith(target,i)]
print("The occurances were seen at positions : ",positions)
```

```python
''' Q6. Write a code to perform basic string compression using the counts of repeated characters '''


word=input("Enter a repeated letter sequence : ")
sorted_word=sorted(word)
result=""
tally=1

for i in range(1,len(sorted_word)):
    if sorted_word[i]==sorted_word[i-1]:
        tally+=1  
    else:
        result+=""+sorted_word[i-1]+str(tally)
        tally=1
result=result+sorted_word[-1]+str(tally)
print(result)
```

```python
'''   BETTER VERSION   '''


temper=input("Enter the arrangement : ")
inv={}

for i in temper :
    if i in inv:
        inv[i]+=1
    else :
        inv[i]=1
print(inv)        
```

```python
''' Q7. Write a code to determine if a string has all unique characters.  '''


word=input("Enter any word of your liking : ")
flag=0
for i in word:
    if word.count(i)==1:
        flag=1
    else :
        flag=0
        break
if flag==0:
    print("The word doesn't have unique characters.")
else :
    print("The word has all unique characters.")
        
```

```python
'''  SMART VERSION  '''

word=input("Enter any word : ")

if len(set(word))==len(word):
    print("All the characters are unique...")
else :
    print("The characters of this word don't appear uniquely...")
```

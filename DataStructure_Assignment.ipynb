```python
'''Q2.  Write a code to count the number of vowels in a string'''

vowels=['a','e','i','o','u']
string = input("Enter any word that contains vowels (preferrably a long one) : ")
count=0
for i in string:
    if i in vowels:
        count+=1
print("The number of vowels in your word is : ",count)        
```

```python
''' Q6. Write a code to perform basic string compression using the counts of repeated characters '''


word = input("Enter a repeated letter sequence: ")


sorted_word = "".join(sorted(word)) 

result = ""
tally = 1

for i in range(1, len(sorted_word)):
    if sorted_word[i] == sorted_word[i-1]:
        tally += 1  
    else:
        result += sorted_word[i-1] + str(tally)
        tally = 1

result += sorted_word[-1] + str(tally)

print(result)
```

```python
''' Q7. Write a code to determine if a string has all unique characters.  '''


word=input("Enter any word of your liking : ")
flag=0
for i in word:
    if word.count(i)==1:
        flag=1
    else :
        flag=0
        break
if flag==0:
    print("The word doesn't have unique characters.")
else :
    print("The word has all unique characters.")
        
```

```python
'''  Q14. Write a code to reverse a list in-place without using any built-in reverse functions  '''


order=input("Enter the elements").split()
print("Reversed : ",order[::-1])                       # this is not in-place swapping
```

```python
order=input("Enter the elements").split()
left=0
right=len(order)-1
while left<right:
    order[left], order[right] = order[right], order[left]

    left+=1
    right-=1
print("Reversed in-place = ",order)    
```

```python
''' Q15. Implement a code to find and remove duplicates from a list while preserving the original order of 
elements '''


order=input("Enter the elements").split()
clean=[]
seen=set()

for x in order:
    if x not in seen:
        clean.append(x)
        seen.add(x)
print("The clean list is : ",clean)       
```

```python
''' Q16 Create a code to check if a given list is sorted (either in ascending or descending order) or not. '''

'''Noob version'''

first=[1,43,21,32,1]

asc=sorted(first)
dsc=sorted(first,reverse=True)

if first==asc:
    print("Ascending")
elif first==dsc:
    print("Descending")
else:
    print("Not arranged")
```

```python
''' Q16 Create a code to check if a given list is sorted (either in ascending or descending order) or not. '''


first=[1,43,21,32,1]

dsc= all(first[i]>=first[i+1] for i in range(len(first)-1))
asc=all(first[i]<=first[i+1] for i in range(len(first)-1))

if asc:
    print("Ascending")
elif dsc:
    print("Descending")
else:
    print("Not arranged")
```

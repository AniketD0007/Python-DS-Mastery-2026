```python
''' Q6. Write a code to perform basic string compression using the counts of repeated characters '''


word = input("Enter a repeated letter sequence: ")


sorted_word = "".join(sorted(word)) 

result = ""
tally = 1

for i in range(1, len(sorted_word)):
    if sorted_word[i] == sorted_word[i-1]:
        tally += 1  
    else:
        result += sorted_word[i-1] + str(tally)
        tally = 1

result += sorted_word[-1] + str(tally)

print(result)
```

```python
''' Q17 Write a code to merge two sorted lists into a single sorted list. '''


first=[int(i) for i in input("Enter the elements of the first list : ").split()]
second=[int(i) for i in input("Enter the elements of the second list : ").split()]

i=j=0

merged=[]

while i<len(first) and j<len(second):
    if first[i]<=second[j]:
        merged.append(first[i])
        i+=1
    else:
        merged.append(second[j])
        j+=1
merged.extend(first[i:])
merged.extend(second[j:])

print(merged)
```

```python
''' Q20. Write a code to shuffle a given list randomly without using any built-in shuffle functions. '''


import random

my_list=input("Enter any list of items separated by space : ").split()
n=len(my_list)

for i in range(n-1,0,-1):
    j=random.randint(0,i)

    my_list[i],my_list[j]=my_list[j],my_list[i]
print("The shuffled list is : ", my_list)    
```

```python
''' Q25. Create a code that takes a tuple and two integers as input. The function should return a new tuple 
containing elements from the original tuple within the specified range of indices '''


def slice_tuple(tup,i,j):
    if i>j:
        i,j=j,i
    return tup[i:j]

tup1=tuple(int(i) if i.lstrip("-").isdigit() else i for i in input("Enter the elements of the list : ").split())
try:
    x = int(input("Enter starting index: "))
    y = int(input("Enter ending index: "))

    if x<0 or y>len(tup1):
        print(f"the index is out of bounds of (0 to {len(tup1)} )")
    print("The sliced result is : ", slice_tuple(tup1,x,y))
except ValueError:
    print("please enter valid integers for the indices")
```

```python
''' Q29. Write a code that takes a tuple and an element as input. The function should return the count of 
occurrences of the given element in the tuple. '''


def occurance(t,elem):
    count=0
    for i in t:
        if i==elem:
            count+=1
    return count

tup=tuple(input("Enter the elements of the tuple : ").split())
element=input("Enter any element from the tuple : ")

print(f"The occurance of {element} in your tuple is {occurance(tup,element)}")
```

```python
''' Q32. Write a code that takes two dictionaries as input and merges them into a single dictionary. If there are 
common keys, the values should be added together. '''


def get_dict():
    keys=input("Enter the keys : ").split()
    values=input("Enter the values : ").split()

    return {k: int(v) for k,v in zip(keys,values)}

dict1=get_dict()
dict2=get_dict()

result=dict1.copy()

for i in dict2:
    if i in result:
        result[i]+=dict2[i]
    else:
        result[i]=dict2[i]

print("The resultant dictionary is : ",result)
```

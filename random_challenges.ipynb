```python
# write a code where you define a function to take a number from the user and give the sum of the digits of that number in return.

def Sum_of_Digits(x):
    digits=[]
    for i in range(len((str(x)))):
        y=x%10
        x=x//10
        digits.append(y)
    return sum(digits)
x=int(input("Enter any number : ")) 
print(Sum_of_Digits(x))
```

```python
# Using List Comprehension :-

def Sum_of_Digits(x):
    if not x.isdigit():
        print("Please enter a number!")
        break
    return sum([int(d) for d in str(x)])
x=input("Enter any number of your liking : ")
Sum_of_Digits(x)
```

```python
def find_average():
    a = input("Enter the numbers: ")
    b = a.split()
    
    # 1. Clean the data first
    numbers = []
    for x in b:
        if x.isdigit():
            numbers.append(int(x))
        else:
            print(f"Warning: '{x}' is not a number and will be skipped.")

    # 2. Check if we actually have numbers to avoid DivisionByZero
    if len(numbers) == 0:
        print("Average = 0 (No valid numbers provided)")
    else:
        # 3. Use built-in sum() for efficiency (Data Science style!)
        total_sum = sum(numbers)
        average = total_sum / len(numbers)
        print("The average of these numbers is:", average)

find_average()
```

```python
def find_average(numbers):
    valid_nos=[int(x) for x in numbers if str(x).isdigit()]
    if not valid_nos:
        return 0
    else :
        return sum(valid_nos)/len(valid_nos)
find_average()        
```

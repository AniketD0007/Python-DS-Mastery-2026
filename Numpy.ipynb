```python
import numpy as np

m=np.random.randint(1,20,size=20).reshape(4,5)
a=m.max()
b=m.min()

nm= (m - b)/(a-b)
print("Max of normalised : ",nm.max())
print("Min of normalised : ",nm.min())
print("Normalised matrix = ",nm)
```

```python
a=np.arange(20, dtype=float)
nans=np.random.choice(np.arange(20), size= 5 ,replace=False)

a[nans]=np.nan
print("\nThe Array with nan values : ",a)
mask=np.isnan(a)
clean=np.nanmean(a)
a[mask]=clean

print("\nThe Array without nan values : ",a)
```

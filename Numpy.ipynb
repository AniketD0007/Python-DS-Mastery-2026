```python
import numpy as np

m=np.random.randint(1,20,size=20).reshape(4,5)
a=m.max()
b=m.min()

nm= (m - b)/(a-b)
print("Max of normalised : ",nm.max())
print("Min of normalised : ",nm.min())
print("Normalised matrix = ",nm)
```

```python
a=np.arange(20, dtype=float)
nans=np.random.choice(np.arange(20), size= 5 ,replace=False)

a[nans]=np.nan
print("\nThe Array with nan values : ",a)
mask=np.isnan(a)
clean=np.nanmean(a)
a[mask]=clean

print("\nThe Array without nan values : ",a)
```

```python
x=np.random.randint(0,100,size=50)
mean=np.mean(x)
std_dev=np.std(x)

z=(x-mean)/std_dev

print("The mean and std.dev before transformation : ",mean," ,",std_dev)
print("The mean and std.dev after transformation : ",np.mean(z).round(6)," ,",np.std(z).round(6))
```

```python
arr=np.append(np.random.randint(0,100,100),[456,-360])
Q1=np.percentile(arr,25)
Q3=np.percentile(arr,75)
IQR=Q3-Q1

LB=Q1-1.5*IQR
UB=Q3+1.5*IQR

Outlier_indices=np.where((arr<LB)|(arr>UB))[0]

print("Q1 : ",Q1,"Q3 : ",Q3,"IQR : ",IQR)
print("Upper Bound : ",UB ,"&","Lower Bound : ",LB)
print("Outliers : ", Outlier_indices)
print("Outlier Values : ",arr[Outlier_indices])
```
